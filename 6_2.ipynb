<h2>Solution</h2>

```python
import pandas as pd
import matplotlib.pyplot as plt

def func(vals) :
    val = list(vals)
    return val

url='https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv'
titanic = pd.read_csv(url)
grouped_titanic = titanic.groupby('sex').agg(func)
female_number = len(grouped_titanic['pclass'].values[0])
male_number = len(grouped_titanic['pclass'].values[1])
vals = [male_number,female_number]
labels = ['Male','Female']
plt.pie(vals,labels=labels,colors=["red","blue"])
plt.title("Male-Female Proportion")
plt.show()
x_female = grouped_titanic['fare'].values[0]
y_female = grouped_titanic['age'].values[0]
x_male = grouped_titanic['fare'].values[1]
y_male = grouped_titanic['age'].values[1]
plt.scatter(x_male,y_male,c="red")
plt.scatter(x_female,y_female,c="blue")
plt.title("Fare vs. Age by Gender")
plt.xlabel("Fare")
plt.ylabel("Age")
plt.show()
```
